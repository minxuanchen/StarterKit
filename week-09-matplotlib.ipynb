```python
# Slide 11
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, normed=1, color='g', alpha=.75)
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
y = matplotlib.mlab.normpdf(bins, mu, sigma)
plt.plot(bins, y, 'k--', linewidth=3.5, color='orange')
plt.annotate('theoretical dist', xy=(110, .020), xytext=(130, .025),arrowprops=dict(facecolor='cyan', shrink=0.3))
plt.show()
```
