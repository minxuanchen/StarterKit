```python
import urllib.request
from bs4 import BeautifulSoup as bs
```

```python
url = "https://en.wikipedia.org/wiki/List_of_rivers_by_discharge"
request = urllib.request.Request(url)
response = urllib.request.urlopen(request)
data = response.read()
response.close()
# Create the soup
soup = bs(data, "html.parser")
table = soup.findAll("table", {"class" : "wikitable"})
print(len(table))
for tables in table:
    rows = tables.findAll('tr')
    r = []
    print(len(rows))
    for row in rows:
        data = row.findAll('td')
        c = []
        for eachdata in data:
            c.append(eachdata)
        r.append(c)
print(r)
```

```python
import pandas as pd
```

```python
pd.DataFrame(r).to_csv("G40116958.csv",ignore_index=True)
```
